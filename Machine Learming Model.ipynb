```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import warnings
warnings.filterwarnings('ignore')
```

```python
df=load_iris()                 ### Loading the iris data from the datasets
```

```python
df
```

```python
features=pd.DataFrame(df.data,columns=df.feature_names)
```

```python
features
```

```python
target=pd.DataFrame(df.target,columns=['target'])
```

```python
target
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
```

### Splitting the model

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,train_size=0.75,random_state=55)    
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

### Building the Model

```python
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
```

```python
model=BaggingClassifier(estimator=RandomForestClassifier(),n_estimators=100,bootstrap=False,max_features=1.0,max_samples=0.9)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred          #### prediction values
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)          ### Checking the accuracy score
```

```python
print(classification_report(y_test,y_pred))  ## checking the precision,recall,f1-score and support
```

```python
cross_val_score(BaggingClassifier(RandomForestClassifier()),features,target,scoring='accuracy',cv=5).mean()
```
